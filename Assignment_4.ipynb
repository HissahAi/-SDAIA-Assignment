# Assignment 4 – Semantic Search for Course Recommendations

## Objective
The goal of this project is to build a semantic search system that recommends courses based on the meaning of the user's query.

## Method
- Create a dataset of course descriptions
- Convert course descriptions into embeddings
- Convert the user query into an embedding
- Use cosine similarity to find the most relevant courses

## Result
The system recommends the courses that are closest in meaning to the user query, even if the wording is different.

```python
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
print("Loading model...")
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded successfully.")
```

```python
courses = [
    "Introduction to Machine Learning: Learn how machines learn from data.",
    "Deep Learning Fundamentals: Neural networks and advanced AI models.",
    "Data Analysis with Python: Analyze and visualize data using Python.",
    "Web Development Bootcamp: Build modern websites using HTML, CSS, and JavaScript.",
    "Cybersecurity Basics: Learn how to protect systems from cyber attacks.",
    "Cloud Computing with AWS: Deploy scalable applications in the cloud.",
    "Natural Language Processing: Build systems that understand human language."
]
```

```python
course_embeddings = model.encode(courses)
print("Course embeddings created successfully.")
```

```python
def search_courses(query, courses, embeddings, top_k=3):
    query_embedding = model.encode([query])[0]

    similarities = np.dot(embeddings, query_embedding) / (
        np.linalg.norm(embeddings, axis=1) * np.linalg.norm(query_embedding)
    )

    top_indices = np.argsort(similarities)[::-1][:top_k]

    results = []
    for idx in top_indices:
        results.append((courses[idx], similarities[idx]))

    return results
```

```python
query = "I want to learn how to train AI models"

results = search_courses(query, courses, course_embeddings)

print("Query:", query)
print()

for i, (course, score) in enumerate(results, 1):
    print(f"Top {i} Course: {course}")
    print(f"Similarity Score: {score:.4f}")
    print()
```

```python
queries = [
    "How can I learn machine learning?",
    "I want to build websites",
    "How do I protect computer systems from hackers?",
    "How do I analyze data with Python?"
]

for q in queries:
    print("=" * 60)
    print("Query:", q)

    results = search_courses(q, courses, course_embeddings)

    for i, (course, score) in enumerate(results, 1):
        print(f"Top {i} Course: {course}")
        print(f"Similarity Score: {score:.4f}")

    print()
```

## Result Analysis

The semantic search system worked successfully and returned the most relevant courses for each query based on meaning.

For example:
- Queries about machine learning returned AI-related courses.
- Queries about websites returned the web development course.
- Queries about security returned the cybersecurity course.
- Queries about data analysis returned the Python data analysis course.

This shows that semantic search can understand the meaning of the query even when the exact words are different.

Using embeddings and cosine similarity makes the search more flexible and more accurate than simple keyword matching.
